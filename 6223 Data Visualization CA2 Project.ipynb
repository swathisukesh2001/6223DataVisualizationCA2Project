# BI-VARIATE


```python
import matplotlib.pyplot as plt
import pandas as pd
titanic_df=pd.read_csv("/content/train.csv")
plt.figure(figsize=(8, 6))
plt.scatter(titanic_df['Age'], titanic_df['Fare'], alpha=0.5)
plt.xlabel('Age')
plt.ylabel('Fare')
plt.title('Age vs. Fare')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
titanic_df=pd.read_csv("/content/train.csv")
plt.figure(figsize=(8, 6))
sns.boxplot(x='Survived', y='Age', data=titanic_df)
plt.xlabel('Survived')
plt.ylabel('Age')
plt.title('Age vs. Survival')
plt.xticks([0, 1], ['Not Survived', 'Survived'])
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
titanic_df=pd.read_csv("/content/train.csv")
plt.figure(figsize=(8, 6))
sns.boxplot(x='Survived', y='Fare', data=titanic_df)
plt.xlabel('Survived')
plt.ylabel('Fare')
plt.title('Fare vs. Survival')
plt.xticks([0, 1], ['Not Survived', 'Survived'])
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
titanic_df=pd.read_csv("/content/train.csv")
class_survival = titanic_df.groupby(['Pclass', 'Survived']).size().unstack()
class_survival.plot(kind='bar', stacked=True)
plt.xlabel('Passenger Class')
plt.ylabel('Count')
plt.title('Pclass vs. Survival')
plt.legend(['Not Survived', 'Survived'], title='Survival')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
titanic_df=pd.read_csv("/content/train.csv")
sns.countplot(x='Sex', hue='Survived', data=titanic_df)
plt.xlabel('Gender')
plt.ylabel('Count')
plt.title('Gender vs. Survival')
plt.legend(['Not Survived', 'Survived'], title='Survival')
plt.show()
```

# UNI-VARIATE


```python
titanic_df=pd.read_csv("/content/train.csv")
survival_counts = titanic_df['Survived'].value_counts()
labels = ['Not Survived', 'Survived']
plt.pie(survival_counts, labels=labels, autopct='%1.1f%%', startangle=90)
plt.title('Survival Distribution')
plt.show()
```

```python
import matplotlib.pyplot as plt
titanic_df=pd.read_csv("/content/train.csv")
gender_counts = titanic_df['Sex'].value_counts()
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', startangle=140)
plt.axis('equal')
plt.title('Gender Distribution')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
titanic_df=pd.read_csv("/content/train.csv")
plt.figure(figsize=(8, 6))
sns.histplot(titanic_df['Age'], kde=True, color='skyblue')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Age Distribution')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
titanic_df=pd.read_csv("/content/train.csv")
class_counts = titanic_df['Pclass'].value_counts()

sns.barplot(x=class_counts.index, y=class_counts.values)
plt.xlabel('Passenger Class')
plt.ylabel('Count')
plt.title('Passenger Class Distribution')
plt.show()
```

```python
titanic_df=pd.read_csv("/content/test.csv")
plt.hist(titanic_df['Fare'], bins=20, edgecolor='k')
plt.title('Fare Distribution')
plt.xlabel('Fare')
plt.ylabel('Frequency')
plt.show()
```
